```python
# now join rating with books
```

```python
rating_with_books = ratings.merge(books, on="isbn")
```

```python
rating_with_books.head()
```

```python
number_of_ratings = rating_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_of_ratings
```

```python
number_of_ratings.head()
```

```python
number_of_ratings.rename(columns={"rating":"num_of_ratings"}, inplace=True)
```

```python
final_ratings = rating_with_books.merge(number_of_ratings, on="title")
```

```python
final_ratings.head()
```

```python
# those books who got atlest 50 ratings
final_ratings = final_ratings[final_ratings['num_of_ratings'] >= 50]
final_ratings
```

```python
final_ratings.head()
```

```python
# lets drop the duplicates
final_ratings.drop_duplicates(['user_id', 'title'], inplace=True)
```

```python
final_ratings.isnull().sum()
```

```python
#  create a pivot table  --> craeting a pivot table for clustering is so important to show collalborative filteing

book_pivot = final_ratings.pivot_table(index='title', columns='user_id', values='rating')
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
#  convert pivot table to csr matrix/sparse matrix
#  sparse matrix is a matrix in which most of the elements are zero

#  for the modal nearest neighbour we use the csr metrice

from scipy.sparse import csr_matrix
```

```python
books_sparse = csr_matrix(book_pivot)
books_sparse
```

```python
type(books_sparse)
```

```python
# lets find the nearest neighbour for the first book
#  we will use brute force algorithm to find the nearest neighbour
#  brute force algorithm is a simple algorithm that finds the nearest neighbour by checking all the points in the dataset
#  brute force algorithm is not efficient for large datasets
from sklearn.neighbors import NearestNeighbors  
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(books_sparse)
```

my training is done now we will focus on testing part --> whether it will show or not

testing model

```python
def recommend_books(book_name):
    ids = np.where(book_pivot.index == book_name)[0][0]
    distances, suggestion = model.kneighbors(book_pivot.iloc[ids,:].values.reshape(1,-1), n_neighbors=6)
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            if j == book_name:
                print(f"you searched'{book_name}\n")
                print(f"The suggestion books are:\n")
            else:
                print(f"{j}\n")
                        
```

```python
book_name = "2nd Chance"
recommend_books(book_name)
```
